```python
#########
import matplotlib
import numpy as np
import matplotlib.pylab as plt
import matplotlib.image as mpimg
import cmath as cm
from scipy.io import wavfile
from scipy import misc 
from scipy import stats
from IPython.display import display
from IPython.display import HTML
import IPython.core.display as di

R_star=np.linspace(0.8,10,10) #in Sun Radius
R_plan=np.linspace(1,500,100) #in Earth Radius

###### change to jovian radii (source: https://en.wikipedia.org/wiki/Jupiter_radius)

R_star_Jov = R_star*9.735
R_plan_Jov = R_plan/11.209 

#print ( len(R_plan_Jov) ,len(R_star_Jov))



plt.figure(figsize=(10,10))
deltas=[]
for i in range(len(R_star_Jov)):
    
    deltas.append((R_plan_Jov/R_star_Jov[i])**2)
    plt.plot(R_plan_Jov,deltas[i],'o--',label=R_star_Jov[i])


plt.ylabel("delta")
plt.xlabel("R_planet")
plt.ylim(0,1)    
plt.legend(title="$R_{*} [R_{J}]$")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

x = np.linspace(0.8,10,10)*9.735 #R_star
y = np.linspace(1,500,100)/11.209  #R_planet
X, Y = np.meshgrid(x, y)
Z = (Y*Y)/(X*X)

fig = plt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.Blues,
                       linewidth=0, antialiased=False)
ax.set_title('surface plot of transit depths')
ax.set_xlabel('R_sun')
ax.set_ylabel('R_planet')
ax.set_zlabel('$\delta$')
fig.colorbar(surf, shrink=0.3, aspect=10)
plt.show()
```

# Exoplanet characterization

### Question 1

Which planet parameters can be derived from radial velocity data? Which parameters cannot?

### Answers

RV cannot determine: Orbit inclination, planet radius.

RV can determine: Planet’s minimum possible mass, semimajor axis of the orbit, period of the orbit, eccentricity of the orbit.

### Question 2

Which planet parameters can be derived from transit data? Which parameters cannot?

### Answers

Transits cannot determine: Planet mass (not even minimum), semimajor axis and eccentricity (Usually it’s assumed a circular orbit whose radius can be easily determined, e=0)

Transits can determine: Planet radius (if star radius is known), star density, period of the orbit, orbit inclination.


### Question 3

Why would it be useful to combine transit and radial velocity observations? Which observing
strategy would you use to do this?

### Answers

Being able to apply both transit and VR methods on the same system would be extremely useful, since it allows completing all the orbital parameters and planet’s main features (Mass, Radius, etc.), this data set would be incomplete as seen previously if only one of the methods is used. Also it increases the precision of the values obtained for, e.g., the planet’s mass, since it would be directly the right mass and not the minimum mass. 

A nice strategy for doing this when possible, is looking for systems that present transits, and when detecting a transit go for a RV measurement, with all the advantages seen previously.


### Question 4

 51 Peg b is the first planet discovered on a main-sequence star. But can we be sure it is truly a
planet? Its RV amplitude is ≈ 55 m/s and it has an orbital period of 4.2 days. What is the minimum
mass? Could it be a brown dwarf instead? If yes, how can this possibility be ruled out? What other
explanations could there be for the observed RV signal?

### Answers

$K=55 m/s$, $P=4.2 days$.
#### Minimum mass:
We know the expression for K:
$$ K = \frac{28.4329}{\sqrt{1-e^{2}}}\left (\frac{M_{p}sin(i)}{M_{J}} \right )\left ( \frac{M_{p}+M_{s}}{M_{\odot}} \right )^{-\frac{2}{3}}\left ( \frac{P}{1yr} \right )^{-1/3} \left  [ \frac{m}{s} \right ] $$

Solving for $ M_p \sin i=M_{Pmin}$, we get:
$$M_{Pmin} = K\frac{\sqrt{1-e^{2}}}{28.4329}\left ( \frac{M_{p}+M_{s}}{M_{\odot}} \right )^{\frac{-2}{3}}\left ( \frac{P}{1yr} \right )^{\frac{-1}{3}} M_{J}$$

Assuming circular orbit $e \rightarrow0$, $M_p+M_s \approx Ms$,  and also taking $M_s=1.06$ M(https://es.wikipedia.org/wiki/Dimidio):
$$M_{Pmin} = K\frac1{28.4329}\left ( \frac{M_{s}}{M_{\odot}} \right )^{\frac{2}{3}}\left ( \frac{P}{1yr} \right )^{\frac{1}{3}} M_{J}
$$
$$M_{Pmin} = K\frac1{28.4329}\left ( 1.06 \right )^{\frac{2}{3}}\left ( \frac{4.2d}{365.25d} \right )^{\frac{1}{3}} M_{J}$$
$$M_{Pmin}=0.45 M_{Jup}$$
If it was a Brown Dwarf $M_P=13 M_{Jup}$  $\sin i = \frac{0.45}{13}, i\approx 2 $°, which is statistically very very unlikely, since the probability distribution is way much denser approaching i=90°, and less denser towards i=0°.
We think a Jupiter-size like planet is the most likely explanation for the RV signal, since other possibilities as Brown Dwarfs are unlikely, star’s intrinsic cycles would not affect its motion at this point of stellar evolution and a binary star would contribute brightness to the system and be detected by direct imaging.


### Bonus 1: 

Using the same expression for K as before, the data given in the excercise and the values of excentricity and mass for Jupiter, Earth and The sun, we get a radial velocity of the star due an Earth-like planet of 0.077488 m/s and a radial velocity of 10.7611 m/s for the jupiter like planet from our point of view. These velocities can be reinterpreted using the Doppler effect formula:

$$\frac{v}{c}=\frac{\Delta \lambda }{\lambda } $$

So the shift in the wavelength would be give by

$$\Delta \lambda=\frac{v}{c}\lambda$$
 
$\lambda$ being the place in the light spectrum where the iron absorption line appears (5000 angstroms ), so the shift for the earth-like planet and the jupiter-like planet would be around 0.001291 angstroms and 0.179352 angstroms accordingly
